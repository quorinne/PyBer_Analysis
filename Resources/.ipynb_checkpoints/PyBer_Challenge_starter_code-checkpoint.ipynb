# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "city_data.csv"
ride_data_to_load = "ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Challenge Deliverable 1. Generate a Ride-Sharing DataFrame by City Type

```python
#  1. Get the total rides for each city type
totalrides = pyber_data_df.groupby(['type']).count()['ride_id']
totalrides
```

```python
# 2. Get the total drivers for each city type
totaldrivers = city_data_df.groupby(['type']).sum()['driver_count']
totaldrivers
```

```python
#  3. Get the total amount of fares for each city type
totalfare = pyber_data_df.groupby(['type']).sum()['fare']
totalfare
```

```python
#  4. Get the average fare per ride for each city type. 
avg_per_ride = totalfare/totalrides
avg_per_ride
```

```python
# 5. Get the average fare per driver for each city type. 
avg_per_driver = totalfare/totaldrivers
avg_per_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
typesum_df = pd.DataFrame({
    'Total Rides': totalrides, 
    'Total Drivers': totaldrivers, 
    'Total Fares': totalfare, 
    'Average Fare Per Ride': avg_per_ride,
    'Average Fare Per Driver': avg_per_driver})

typesum_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
typesum_df.index.name = None
```

```python
#  8. Format the columns.
typesum_df['Total Rides'] = typesum_df['Total Rides'].map("{:.0f}".format)
typesum_df['Total Drivers'] = typesum_df['Total Drivers'].map("{:.0f}".format)
typesum_df['Total Fares'] = typesum_df['Total Fares'].map("${:,.2f}".format)
typesum_df['Average Fare Per Ride'] = typesum_df['Average Fare Per Ride'].map("${:,.2f}".format)
typesum_df['Average Fare Per Driver'] = typesum_df['Average Fare Per Driver'].map("${:,.2f}".format)
typesum_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# Print the merged DataFrame for reference.
pyber_data_df
```

```python
# 1. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.

fare_by_city_date_df = pyber_data_df.groupby(['type','date']).sum().fare.to_frame()
fare_by_city_date_df.head()
```

```python
# 2. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
ctd_df = fare_by_city_date_df.reset_index()
ctd_df.head()
```

```python
# 3. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
ctd_df = ctd_df.pivot(index='date',columns='type', values='fare')
ctd_df.head()
```

```python
ctd_df 
```

```python
# 4. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
dates_df = ctd_df.loc['2019-01-01':'2019-04-29']
dates_df
```

```python
# 5. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)

dates_df.index = pd.to_datetime(dates_df.index)

dates_df.head()

```

```python
# 6. Check that the datatype for the index is datetime using df.info()
dates_df.info
```

```python
# 7. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.


dates_df = dates_df.resample('W').sum()
dates_df


```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

dates_df.plot(figsize=(16,5),)
plt.xlabel('Months')
plt.ylabel('Fare($)')
plt.title('Total Fare by City Type')

plt.savefig("../Analysis/Del2.png")
```
